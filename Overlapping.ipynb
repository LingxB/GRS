```python
def overlap2d(df,catname,welgname):
    
    c_cats = Counter()
    c_welg = Counter()

    for x in df[catname]:
        c_cats.update(x.split(','))

    for x in df[welgname]:
        c_welg.update(x.split(','))

    cats = sorted(c_cats.keys())
    welgs = sorted(c_welg.keys())

    c_matrix = np.zeros([len(cats),len(welgs)],dtype=int)

    for idx,row in df.iterrows():
        cat = row[catname].split(',')
        welg = row[welgname].split(',')
        for c in cat:
            for w in welg:
                c_matrix[cats.index(c),welgs.index(w)] += 1
    
    return cats,welgs,c_matrix

def matrix2df_(cols,rows,matrix):
    df = pd.DataFrame(index=cols,columns=rows,data=matrix)
    return df

def topNoverlap_(df,n=10):
    
    rows = df.index
    cols = df.columns
    tops = []
    
    for idx,row in df.iterrows():
        sortedlst = sorted([(count,idx) for idx,count in enumerate(row)],reverse=True)
        topN = [(cols[idx],count) for count,idx in sortedlst[:n]]
        tops.append(topN)
        
    new_df = pd.DataFrame(index=rows,columns=range(1,n+1),data=tops)
    return new_df
```

```python
cats,welgs,C_mat = overlap2d(df,'CATS','WELLGOS')
```

```python
df_C_mat = matrix2df_(cats,welgs,C_mat)
```

```python
df_C_mat.to_csv('Overlap/CATvsWelg_G-even.csv')
```

```python
top_cw = topNoverlap_(df_C_mat)
```

```python
top_cw.to_csv('Overlap/CATvsWelg_Top10_G-even.csv')
```

```python
def countoverlaps(df,df_=None):
    
    c = Counter()
    for x in df:
        c.update(x.split(','))

    elements = sorted(c.keys())

    c_matrix = np.zeros([len(elements),len(elements)],dtype=int)

    for idx,ele in enumerate(elements):
        for _ele in df:
            _elst = _ele.split(',')
            if ele in _elst:
                co_ocur = [e for e in _elst if e != ele]
                for s in co_ocur:
                    c_matrix[idx,elements.index(s)] += 1

    return elements,c_matrix
        
        
```

```python
def matrix2df(elements,matrix):
    df = pd.DataFrame(index=elements,columns=elements,data=matrix)
    return df
```

```python
welgolst,C_welgo = countoverlaps(df['WELLGOS'])
catlst,C_cats = countoverlaps(df['CATS'])
```

```python
df_C_welgo = matrix2df(welgolst,C_welgo)
df_C_cats = matrix2df(catlst,C_cats)
```

```python
df_C_welgo.to_csv('Overlap/Welgo_Overlap_Matrix_G-even.csv')

df_C_cats.to_csv('Overlap/Cat_Overlap_Matrix_G-even.csv')
```
